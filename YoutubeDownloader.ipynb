```python
import json
import yt_dlp
from youtubesearchpython import ChannelSearch,ResultMode,CustomSearch
from concurrent.futures import ThreadPoolExecutor
```

```python
def download_videos(video):
    # Define the path where videos will be saved
    SAVE_PATH = "/media/aquinate/5CF07692F07671DE/videos_youtube"
    video_error = []
    # Example link of the video to be downloaded (make sure this is a valid YouTube URL)
    link = 'https://www.youtube.com'+video  # Replace this with your actual link
    
    ydl_opts = {
        'format': 'bestvideo+bestaudio/best',
        'outtmpl': f'{SAVE_PATH}/%(title)s.%(ext)s',
    }
    
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([link])
        print('Video downloaded successfully!')
    except Exception as e:
        print(f"Error downloading video: {e}")
        video_error.append(link)

    return video_error
```

```python
def list_videos_channel():
    search = ChannelSearch("gremio+prudente", "UCBFcL3nMWFcNnoWOEtacp1w", language = "pt-br", region = "BR")
    result = search.result(mode = ResultMode.dict)['result']
    youtube_videos = [i['uri'] for i in result]
    return youtube_videos
```

```python
def main():
    # Define o número máximo de threads que você deseja usar
    video_error = []
    max_threads = 10
    youtube_videos = list_videos_channel()
    with ThreadPoolExecutor(max_threads) as executor:
        video_error.append(executor.map(download_videos, youtube_videos))
    return video_error
```

```python
if __name__ == '__main__':
    video_error = main()
```
